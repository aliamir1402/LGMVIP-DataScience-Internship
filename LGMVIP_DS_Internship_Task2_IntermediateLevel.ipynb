```python
#Ali Amir Khawaja 
#Intermediate-2-Prediction using Decision Tree Classifier
import pandas as pd
import numpy as np
import graphviz
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split


# Specify the path to your Excel file
excel_file = ''

# Read the excel file as a pandas DataFrame for different sheets and store them in a list
# Read the Excel file into a pandas DataFrame
data = pd.read_csv('/content/LGMVIP_DS_Internship-Task2-IntermediateLevel.csv')

# Now you can work with the data in the DataFrame
# For example, you can print the first few rows
#print(data)

data=np.array(data)
data=np.transpose(data)
x=np.transpose(data[1:5])
y=np.transpose(data[5:6])
print("Features:\n",x)
print()
print("Expected Output:\n",y)
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100,"%")
```

```python
# Visualize the decision tree
dot_data = export_graphviz(clf, out_file=None, feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'], class_names=['Iris-setosa','Iris-virginica','Iris-versicolor'], filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("decision_tree")
graph.view()
```
