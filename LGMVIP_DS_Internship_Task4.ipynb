```python
#LGMVIP-DS-Internship_Task-4-Ali Amir Khawaja

#Libraries Used is this task
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image_path = '/content/img.png'
image = cv2.imread(image_path)

desired_width = 500
desired_height = 300

resized_image = cv2.resize(image, (desired_width, desired_height))

cv2_imshow(resized_image)
```

```python
#converting image to greyscale using .cv2Color function
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray_image)
```

```python
# Invert the image
inverted_image = 255 - gray_image

# Display the inverted image
cv2_imshow(inverted_image)
```

```python
# Apply blur effect to the image
blurred_image = cv2.blur(inverted_image, (15, 15))

# Display the blurred image
cv2_imshow(blurred_image)
```

```python
# Get the dimensions of the resized image
height, width, channels = resized_image.shape

# Convert the gray_image and blurred_image to numpy arrays
gray_image = np.array(gray_image)
blurred_image = np.array(blurred_image)

# Iterate through each pixel in the image
for i in range(height-1):
    for j in range(width-1):
        # If the pixel in the gray image is 0, set it to 1
        if gray_image[i][j] == 0:
            gray_image[i][j] = 1

        # If the pixel in the blurred image is 0, set it to 1
        if blurred_image[i][j] == 0:
            blurred_image[i][j] = 1
```

```python
# Divide the gray image by the blurred image element-wise
pencil_image = gray_image - blurred_image 

# Display the resulting pencil image
cv2_imshow(pencil_image)
```
